```python
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', 2000)
pd.set_option('display.max_columns', None)
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pupil_data = pd.read_csv("../experiments/nancy_sus/smoothed/p1.csv", usecols=['SAMPLE_INDEX','TRIAL_INDEX','SMOOTHED','STIMULI','SYSTEMS','BLOCKS','SUBJECT'])
```

```python
pupil_data.head()
```

```python
aggregate = pupil_data.groupby(['SYSTEMS','SAMPLE_INDEX'])['SMOOTHED'].mean().reset_index()
aggregate
```

```python
system_names = aggregate.SYSTEMS.unique()
sampling_freq = 50

for system in system_names:
    systems_data = aggregate[aggregate['SYSTEMS'] == system].copy()
    ax = systems_data['SMOOTHED'].plot(use_index=False, label=system, legend=True)
   
```
